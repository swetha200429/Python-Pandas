```python
import pandas as pd
df=pd.read_csv("orders_wrong.csv")
print("Original Data\n")
print(df)
df.loc[df["Price"]<0,"Price"]=abs(df["Price"])
df.loc[df["Quantity"]==0,"Quantity"]=1
df["City"]=df["City"].replace("Chennai","Chennai")
df.loc[df["Price"]>100000,"Price"]=10000
print("\nCleaned Data\n")
print(df)
```

```python
import pandas as pd
df=pd.read_csv("event.csv")
print("Original Data\n")
print(df)
dup=df.duplicated()
print("\nDuplicate rows\n")
print(df[dup])
count=dup.sum()
df_clean=df.drop_duplicates()
print("\nCleaned Data\n")
print(df_clean)
print("\nNumber of duplicates removed:",count)
```
