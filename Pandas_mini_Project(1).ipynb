```python
import pandas as pd
data={
    "Name":["Inception","Avatar","Titanic","Joker","Interstellar"],
    "Year":[2010,2009,1997,2019,2014],
    "Rating":[8.8,7.8,7.9,8.5,8.6],
    "Duration":[153,735,734,865,242]
}
df=pd.DataFrame(data)
print("Full DataFrame\n")
print(df)
print("\nMovie Names\n")
print(df["Name"])
print("\nSecond movie\n")
print(df.loc[1])
print("\nRating>8\n")
print(df[df["Rating"]>8])
print("\nFirst 3 rows\n")
print(df.head(3))
```
