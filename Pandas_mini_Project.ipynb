```python
import pandas as pd
names=pd.Series(["Rice","Milk","Bread","Eggs","Sugar"])
price=pd.Series([50,53,87,12,90])
qty=pd.Series([10,20,15,12,18])
df=pd.DataFrame({
    "Name":names,
    "Price":price,
    "Qty":qty
})
print("Full DataFrame\n")
print(df)
print("\nPrice column\n")
print(df["Price"])
print("\nFirst 2 rows\n")
print(df.head(2))
```

```python
import pandas as pd
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
water=[2,3,2.5,3.5,4,3,2]
s=pd.Series(water,index=days)
print("Full Series\n")
print(s)
print("\nWednesday intake:",s["Wed"])
print("\nFirst 3 days\n")
print(s[:3])
print("\nMax intake:",s.max())
```
